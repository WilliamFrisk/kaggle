```python
# Load data
raw_train_df = pd.read_csv('../data/raw/train.csv')
raw_test_df = pd.read_csv('../data/raw/test.csv')
```

# Data Preprocessing
02_data_preprocessing.ipynb

```python
def fix_na(df):
    X = df.copy()
    X['Median_Episode_Length_minutes'] = (X
                            .groupby('Podcast_Name')['Episode_Length_minutes']
                            .transform('median'))
    X['Episode_Length_minutes'] = X['Episode_Length_minutes'].fillna(X.pop('Median_Episode_Length_minutes'))
    X['Number_of_Ads'] = X['Number_of_Ads'].fillna(0)
    X['Guest_Popularity_percentage'] = X['Guest_Popularity_percentage'].fillna(1)
    X['Episode_Title'] = X['Episode_Title'].str.split().str[1].astype(int)

    return X

X_train = fix_na(raw_train_df)
y_train = X_train.pop('Listening_Time_minutes')

X_test = fix_na(raw_test_df)

X_train.head()
```

### Storing

```python
X_train.to_csv('../data/processed/X_train.csv', index=False)
X_test.to_csv('../data/processed/X_test.csv', index=False)
y_train.to_csv('../data/processed/y_train.csv', index=False)
```
