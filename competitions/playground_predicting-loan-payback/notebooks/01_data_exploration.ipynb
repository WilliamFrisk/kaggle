```python
column_types = pd.DataFrame({
    'Column': raw_train_df.columns,
    'Type': ['Numerical' if pd.api.types.is_numeric_dtype(raw_train_df[col]) else 'Categorical' for col in raw_train_df.columns]
})

print(column_types)
```

```python
# Filter categorical columns
categorical_cols = [col for col in raw_train_df.columns if not pd.api.types.is_numeric_dtype(raw_train_df[col])]

# Display unique values for each categorical column
for col in categorical_cols:
    print(f"{col}: {raw_train_df[col].unique()}, num unique: {len(raw_train_df[col].unique())}")
```
