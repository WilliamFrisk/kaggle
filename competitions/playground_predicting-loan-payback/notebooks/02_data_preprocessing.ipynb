```python
# Load data
raw_train_df = pd.read_csv('../data/raw/train.csv')
raw_test_df = pd.read_csv('../data/raw/test.csv')
```

# Data Preprocessing
02_data_preprocessing.ipynb

```python
train_df = raw_train_df.copy()

train_df.head()
```

## One-hot encoding

```python
train_df = pd.get_dummies(train_df, columns=['gender', 'marital_status', 'education_level', 'employment_status', 'loan_purpose'], dtype=float)

indices, categories = pd.factorize(train_df['grade_subgrade'])
grade_mapping = dict(zip(train_df['grade_subgrade'], indices))
train_df['grade_subgrade'] = train_df['grade_subgrade'].map(grade_mapping).fillna(-1.0).astype(float)

train_df.head()
```

```python
test_df = pd.get_dummies(raw_test_df, columns=['gender', 'marital_status', 'education_level', 'employment_status', 'loan_purpose'], dtype=float)
test_df['grade_subgrade'] = test_df['grade_subgrade'].map(grade_mapping).fillna(-1.0).astype(float)


test_df.head()
```

```python
train_df.to_csv('../data/processed/train.csv', index=False)
test_df.to_csv('../data/processed/test.csv', index=False)
```
